```python
import seaborn as sns
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris =sns.load_dataset("iris")
iris
```

```python
iris.loc[0:10,'sepal_length']=np.nan
```

```python
iris.isnull().sum()
```

```python
iris['sepal_length'].fillna(iris['sepal_length'].mean(),inplace=True)
```

```python
iris.isnull().sum()
```

```python
iris.info()
```

```python
iris.drop_duplicates(inplace=True)
```

```python
iris.info()
```

```python
feature=['sepal_length','sepal_width','petal_length','petal_width']
for i in feature:
    print(f"Mean of {i} = {np.mean(iris[i]):.2f}","\t"
          f"Median of {i} = {np.median(iris[i]):.2f}","\t"
          f"Variance of {i} = {np.var(iris[i]):.2f}")
```

```python
iris['species'].unique()
```

```python
#Filtering rows
Setosa = iris[iris['species']=='setosa']
Setosa
```

```python
Versicolor = iris[iris['species']=='versicolor']
Versicolor
```

```python
Virginica = iris[iris['species']=='virginica']
Virginica
```

```python
species_average=iris.groupby('species').mean().round(2)
species_average
```

```python
plt.hist(iris['sepal_length'],bins=10)
plt.title('Histogram of sepal length')
plt.show()
```

```python
fig=plt.figure(figsize=(10,4))
sns.scatterplot(x='sepal_length',y='petal_length',hue='species',data=iris,marker='*',s=100)
plt.title('Sepal Length vs Petal Length')
plt.show()               
```

```python
fig=plt.figure(figsize=(20,15))
for i in range(len(feature)):
    plt.subplot(2,2,i+1)
    sns.boxplot(x='species',y=feature[i],data=iris)
    plt.title(f"Boxplot of {feature[i]} by Species")
plt.show
```

```python
dates = pd.date_range(start='2025-01-01', periods=len(iris))
data = iris.copy()
data.index = dates
window_size = 10
rolling_data = data[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']].rolling(window=window_size).mean()

plt.figure(figsize=(12,10))

for i in range(len(feature)):
    plt.subplot(2,2,i+1)
    plt.plot(data.index, data[feature[i]], color="red", label=f"{feature[i]}")   # raw data
    plt.plot(rolling_data.index, rolling_data[feature[i]], color="black", label=f"{feature[i]} (rolling avg)")  # rolling average
    plt.title(f"{feature[i]} with Rolling Average")
    plt.legend()

plt.tight_layout(pad=4)
plt.show()  
```
